```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import HDBSCAN
hdbscan = HDBSCAN(min_cluster_size=5, metric='euclidean', algorithm='brute', cluster_selection_method='leaf').fit(X)
hdbscan
```

```python
hdbscan.labels_
```

```python
hdbscan.n_features_in_
```

```python
hdbscan.probabilities_
```

```python
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=hdbscan.labels_
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, facet_kws={"legend_out": True})
```
