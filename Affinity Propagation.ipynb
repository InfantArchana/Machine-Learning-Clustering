```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(random_state=5)
y_clustering = clustering.fit_predict(X)
clustering
```

```python
clustering.labels_
```

```python
len(clustering.cluster_centers_)
```

```python
clustering.cluster_centers_
```

```python
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=clustering.labels_
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, facet_kws={"legend_out": True})
```
