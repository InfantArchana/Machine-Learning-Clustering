```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import SpectralClustering
sc = SpectralClustering(n_clusters=5, assign_labels='discretize', affinity='nearest_neighbors', random_state=0)
y_sc = sc.fit_predict(X)
```

```python
sc
```

```python
sc.labels_
```

```python
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=sc.labels_
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, facet_kws={"legend_out": True})
```
